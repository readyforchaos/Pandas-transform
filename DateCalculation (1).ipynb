```python
#Parse the Date1 and Date2 columns into proper datetime
dateparse = lambda dates: pd.datetime.strptime(dates, '%Y%m%d')
df = pd.read_csv('test.csv', parse_dates=['Date1', 'Date2'], date_parser=dateparse)
```

```python
df
```

```python
#Create a new column called 'NumberOfDays'
df['NumberOfDays'] = (df.Date2 - df.Date1)
```
